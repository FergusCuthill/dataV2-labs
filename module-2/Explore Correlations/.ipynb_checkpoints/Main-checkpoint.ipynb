```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
cdd = pd.read_csv("CDD.csv", sep="	")
columns_0s = ["MAGNITUDE","INJURED / INFECTED", "EVACUATED", "UTILITY - PEOPLE AFFECTED"]
cdd[columns_0s] = cdd[columns_0s].replace(0, np.nan)
cdd
```

```python
cdd.corr()
```

```python
sns.heatmap(cdd.corr())
```

```python
"""
Positive correlation - NGO payments and Injured/Infected. This is likely because the NGOs will need to do a lot of work to 
help the victims of a disaster. 
No correlation - Provincial department payments and fatalities. I suspect here, that provincial departments do not deal 
with payments relating to fatalities. 
Negative correlation - Magnitude and OGD costs. I suspect there are only two values here which allows a perfect line of 
best fit. 
"""
```
