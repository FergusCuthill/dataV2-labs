#### Read the `employees` dataset from Ironhack's database:

#### What's the mean `Salary` by `Department`?
**Tip**: There are 2 ways to do it: with or without pivot tables. Do it both ways. 

```python
# Method 1
# your code here

emps.groupby("Department").mean()["Salary"]
```

```python
# Method 2

emps.pivot_table(index="Department", values="Salary")
```

#### Now, group the data by more than one column. Use Pandas pivot table function to select the mean `Salary` by `Department` and `Title`.

```python
# your code here
emps.pivot_table(index="Department", columns="Title", values="Salary")
```

#### Display the number of `Years` of work experience and the `Salary` for each `Title`. Add `Years` as a column in your pivot table.
**Hint**: Use Pandas documentation to explore the [pivot table](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) function.

```python
# your code here
emps.pivot_table(index="Title", columns="Years", values="Salary")
```

#### Find the mean `Salary` and number of employees in each `Department` and for each job `Title`.

```python
# your code here
emps.pivot_table(index="Department", columns="Title", values=["Salary"], aggfunc=["mean","count"])
```

#### For each `Department` and each job `Title`, find the `Salary` median, minimum, maximum and standard deviation. Comment your results.

```python
# your code here
emps.pivot_table(index="Department", columns="Title", values=["Salary"], 
                 aggfunc=["median", "min", "max", np.std])
```

```python
"""
Some departments do not have an employee with a certain title and so these are returned as NaN. 
VP and analyst pay does not seem to shift between departments while associate does slightly. 

"""
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# your code here
# The NaN values represent combinations of department and title that do not exist therefore a dash seems appropriate
emps.pivot_table(index="Department", columns="Title", values=["Salary"], 
                 aggfunc=[np.median, np.min, np.max, np.std], fill_value="-")
```

#### The stake holders want to know, for each `Department`, the number of employees and how much money is spent on salaries. Could you provide that information? 

```python
# your code here
emps.pivot_table(index="Department", values="Salary", aggfunc=["count","sum"])
```

#### For each `Department` and `Title`, find the the total years of work experience and the mean `Salary`.

```python
# your code here
emps.pivot_table(index="Department", columns="Title", values=["Years","Salary"], aggfunc={"Years":"sum", "Salary":"mean"})
```

## Bonus Challenge 1

#### Compute the mean `Salary` after removing the maximum `Salary` for each `Department`. Create the appropriate pivot table.
**Hint**: Write a custom function.

```python
# your code here
emps.pivot_table(index="Department", values="Salary", aggfunc=lambda x: (sum(x) - max(x))/(len(x)-1))
```
