#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
# This code works but it's weird to me that it does so given that I call the log scaling on p outside the for loop.
# I would assume that once the for loop starts again, new values are assigned to a new plot as p.plot is run a second time
# The log scaling then continues to affect the last plot opened, in this case the second which is what we want. 
# I wasn't very happy with this method though so in the below cells are my attempts to make this work better. 
plots = [ax1, ax2]
fig, plots = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for p in plots:
    p.plot(x,x**2, label='y=x^2')
    p.plot(x,np.exp(x), label='y=e^x')
    p.set_xlabel('x')
    p.set_ylabel('y')
    p.legend()

p.set_yscale("log")
p.set_xscale("log")

fig.tight_layout(pad=3.0)

plt.show()
```

```python
# Just to demonstrate that this performs the same as above

plots = [ax1, ax2]
fig, plots = plt.subplots(1, 2, figsize=(10, 4))

for i in range(len(plots)):
    plots[i].plot(x,x**2, label='y=x^2')
    plots[i].plot(x,np.exp(x), label='y=e^x')
    plots[i].set_xlabel('x')
    plots[i].set_ylabel('y')
    plots[i].legend()

plots[i].set_yscale("log")
plots[i].set_xscale("log")

fig.tight_layout(pad=3.0)

plt.show()
```

```python
# This makes more sense. 

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

for i in range(len(axs)):
    axs[i].plot(x,x**2, label='y=x^2')
    axs[i].plot(x,np.exp(x), label='y=e^x')
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('y')
    axs[i].legend()

axs[1].set_yscale("log")
axs[1].set_xscale("log")

fig.tight_layout(pad=3.0)

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# Set up plots
fig, [ax1,ax2] = plt.subplots(1, 2, figsize=(12, 4))

# MEDIAN STEPS BY MONTH_ENCODED
sbm = (fitbit.pivot_table(index="Months", values=["Months_encoded","Steps"],aggfunc=np.median)
       .sort_values("Months_encoded").reset_index())

ax1.bar(sbm["Months"],sbm["Steps"])
ax1.tick_params(axis="x",rotation=-45)

# MEDIAN STEPS BY WORK_OR_WEEKEND
sbd = fitbit.pivot_table(index="Work_or_Weekend", values="Steps",aggfunc=np.median).reset_index()

ax2.bar(sbd["Work_or_Weekend"].map({0:"Weekend", 1:"Weekday"}),sbd["Steps"])
ax2.tick_params(axis="x",rotation=-45)

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

plots = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]

fig, axs = plt.subplots(1, 3, figsize=(12, 4))

for i in range(len(axs)):
    axs[i].scatter(fitbit[plots[i]],fitbit["Steps"])
    axs[i].set_xlabel(plots[i])
    axs[i].set_ylabel("Steps")

fig.tight_layout(pad=3.0)
plt.show()
```
