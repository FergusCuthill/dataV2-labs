## Developing a Model

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
# Your code here
# bhps = bhp.sample(frac=1).reset_index(drop=True)
from sklearn.model_selection import train_test_split

# Excluding rad because it correlates highly with tax
X_train, X_test, y_train, y_test = train_test_split(bhp.drop(["medv", "rad"], axis=1), bhp["medv"], test_size=0.2)
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
# Five separate RFR here with the given max depths
from sklearn.ensemble import RandomForestRegressor

rfms = [RandomForestRegressor(max_depth=i).fit(X_train, y_train) for i in range(2,11,2)]
```

Now, plot the score for each tree on the training set and on the testing set.

```python
# Produce a plot with the score for the testing and training for the different max depths
rftr = [r2_score(y_train, rfms[i].predict(X_train)) for i in range(5)]
rfte = [r2_score(y_test, rfms[i].predict(X_test)) for i in range(5)]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(range(2,11,2), rftr, label = "train")
ax.plot(range(2,11,2), rfte, label = "test")
ax.legend()
plt.show()
```

What do these results tell you about the effect of the depth of the trees on the performance of the model?

### My comments
This varies a little when code is re-run but generally, having more than about 4 features  doesn't significantly improve the fit on the test data while the training fit will tend to improve suggesting that the model is overfitting. 

### Bias-Variance Tradeoff
When the model is trained with a maximum depth of 1, does the model suffer from high bias or from high variance? How about when the model is trained with a maximum depth of 10?

### My comments
At depth 1, the model has a high bias and low variance. At depth 10, the model has a high variance and low bias. 

### Best-Guess Optimal Model
What is the max_depth parameter that you think would optimize the model? Run your model and explain its performance.

```python
# Your response here
rfms = [RandomForestRegressor(max_depth=i).fit(X_train, y_train) for i in range(2,11)]
rftr = [r2_score(y_train, rfms[i].predict(X_train)) for i in range(9)]
rfte = [r2_score(y_test, rfms[i].predict(X_test)) for i in range(9)]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(range(2,11), rftr, label = "train")
ax.plot(range(2,11), rfte, label = "test")
ax.legend()
plt.show()
```

Every time I run the code, I get slightly different results but usually, at around about 4, the test performance stops improving so I would be inclined to use a value of 4 or 5.
